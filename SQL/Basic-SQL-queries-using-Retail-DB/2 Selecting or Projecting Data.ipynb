```python
%%sql 

SELECT * FROM information_schema.columns 
WHERE table_catalog = 'retail_db' 
    AND table_name = 'orders'
```

```python
%%sql 

SELECT order_customer_id, order_date, order_status 
FROM orders 
LIMIT 10
```

```python
%%sql
SELECT order_customer_id,
    to_char(order_date,'yyyy-MM'), 
    order_status
FROM orders
LIMIT 10;
```

Now using AS to change how the new column is shown. 

```python
%%sql
SELECT order_customer_id,
    to_char(order_date,'yyyy-MM') AS order_month, 
    order_status
FROM orders
LIMIT 10;
```

```python
%sql SELECT DISTINCT to_char(order_date, 'yyyy-MM') AS order_month FROM orders;
```

```python
%sql SELECT count(1) FROM orders
```

```python
%%sql 

SELECT count(DISTINCT to_char(order_date, 'yyyy-MM')) AS distinct_month_count 
FROM orders
```
